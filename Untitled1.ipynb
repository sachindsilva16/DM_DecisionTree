```python
import pandas as pd
```

```python
df=pd.read_csv('Titanic-Dataset.csv')
```

```python
df
```

```python
nec_inputs=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked','Survived'],axis='columns')
```

```python
nec_inputs
```

```python
target=df['Survived']
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_Pclass=LabelEncoder()
label_Sex=LabelEncoder()
label_Age=LabelEncoder()
label_Fare=LabelEncoder()
```

```python
nec_inputs['Pclass_values']=label_Pclass.fit_transform(nec_inputs['Pclass'])
```

```python
nec_inputs
```

```python
nec_inputs['Sex_values']=label_Pclass.fit_transform(nec_inputs['Sex'])
```

```python
nec_inputs
```

```python
nec_inputs['Age_values']=label_Pclass.fit_transform(nec_inputs['Age'])
```

```python
nec_inputs
```

```python
nec_inputs['Fare_values']=label_Pclass.fit_transform(nec_inputs['Fare'])
```

```python
nec_inputs
```

```python
nec_input_values=nec_inputs.drop(['Pclass','Sex','Age','Fare'],axis='columns')
```

```python
nec_input_values
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(nec_input_values,target)
```

```python
model.score(nec_input_values,target)
```
